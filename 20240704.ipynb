```python
df = pd.read_csv('/content/drive/MyDrive/datasets/medical_cost/medical_cost.csv')
df.head()
```

```python
age_bin_list = np.arange(10, 80, 10)
df['age_bin'] = pd.cut(df['age'], bins=age_bin_list)

pivot_df = df.pivot_table(
    index='age_bin', columns='region', values='charges', aggfunc='median'
)
```

```python
fig, ax = plt.subplots()
sns.heatmap(pivot_df, ax=ax, annot=True)
```

```python
fig, ax = plt.subplots()
sns.heatmap(pivot_df, ax=ax, annot=True, fmt='.2e')
```

```python
fig, ax = plt.subplots()
sns.heatmap(
    pivot_df, ax=ax, annot=True, fmt='.2e',
    vmax=16000, vmin=0, cmap='RdBu'
)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

sns.boxplot(x='company_size', y='salary_in_usd', data=df, ax=ax[0], order=['S','M','L'])
ax[0].set_title('company_size box plot', fontsize=16)

sns.histplot(x='salary_in_usd', data=df, ax=ax[1])
ax[1].set_title('salary histogram', fontsize=16)

pivot_df = df.pivot_table(
    index='company_size', columns='experience_level', values='salary_in_usd', aggfunc='mean'
)
sns.heatmap(pivot_df, ax=ax[2], annot=True)
ax[2].set_title('heatmap of\nexperience_level x company_size', fontsize=16)
```
